```python
import json
import matplotlib.pyplot as plt

# Read the JSON file
with open("result.json", "r") as file:
    data = json.load(file)

history = {}

for i, neuron in enumerate(data):
    for key, value in neuron.items():
        if not key in history:
            history[key] = []
        history[key].append(value)
```

```python
plt.plot(history["V_s"])
```
