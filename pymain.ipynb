```python
import json
import matplotlib.pyplot as plt

# Read the JSON file
with open("result.json", "r") as file:
    data = json.load(file)

history = {}

for i, neuron in enumerate(data):
    for key, value in neuron.items():
        if not key in history:
            history[key] = []
        history[key].append(value)
    
plt.plot(history["V"])
plt.show()
```

```python
import json
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation

def load_and_visualize_json(filename):
    # Load JSON data
    with open(filename, 'r') as f:
        data = json.load(f)
    
    # Convert to numpy arrays
    frames = [np.array(frame) for frame in data]
    print(np.max(frames))
    # return
    # Set up the plot
    fig, ax = plt.subplots()
    img = ax.imshow(
        frames[0],
        cmap='viridis',
        vmin=-70,  # Minimum value (black)
        vmax=40    # Maximum value (white)
    )
    plt.colorbar(img, label='V_s')
    ax.set_title("Neuron Grid")

    # Update function for animation
    def update(frame):
        img.set_array(frame)
        return [img]
    
    # Create animation
    ani = FuncAnimation(
        fig,
        update,
        frames=frames,
        interval=10,  # 200ms per frame
        blit=True,
        repeat=True
    )
    
    plt.show()

load_and_visualize_json("grid.json")
```
